```python
from imblearn.over_sampling import RandomOverSampler
os=RandomOverSampler(0.5)
x,y=os.fit_resample(x,y)
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=12)
x=pca.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=40,stratify=y)
```

```python
print(x_train.shape,x_test.shape)
```

```python
print(y_train.shape,y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

#importing error metrics
from sklearn.model_selection import GridSearchCV,cross_val_score
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
model=[GaussianNB(),KNeighborsClassifier()]


for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print("Accuracy Score of",m,'is:')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('*************************************************************************')
    print('\n')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
models=[RandomForestClassifier(),AdaBoostClassifier()]


for m in models:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print("Accuracy Score of",m,'is:')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('\n')
```

```python
rfc=RandomForestClassifier()
parameters={  'bootstrap': [True], 'max_depth': [5, 10, None],
             'n_estimators': [5, 6, 7, 8]}
clf=GridSearchCV(rfc,parameters)
clf.fit(x,y)
clf.best_params_
```

```python
rfc=RandomForestClassifier(bootstrap= True,max_depth= None,n_estimators=7)
rfc.fit(x_train,y_train)
rfc.score(x_train,y_train)
pred_rfc=rfc.predict(x_test)
print("Accuracy Score of",pred_rfc,'is:')
print(accuracy_score(y_test,pred_rfc))
print(confusion_matrix(y_test,pred_rfc))
print(classification_report(y_test,pred_rfc))
print('\n')
```

```python
score=cross_val_score(rfc,x,y,cv=10,scoring='accuracy')
print("Model:",rfc)
print('Score:',score)
print('Mean Score:',score.mean())
print('Standard Deviation ',score.std())
print('*************************************************************************')
print('\n')
```
